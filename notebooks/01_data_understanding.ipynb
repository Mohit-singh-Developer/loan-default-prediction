## going to visualize

```python
import matplotlib.pyplot as plt

df["AMT_INCOME_TOTAL"].hist(bins=50)

plt.title("Distribution of Annual Income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
df["AMT_CREDIT"].hist(bins=50)

plt.title("Distribution of Credit Amount")
plt.xlabel("Credit Amount")
plt.ylabel("Frequency")
plt.show()
```

### box plot

```python
plt.figure(figsize=(10,4))
plt.boxplot(df["AMT_INCOME_TOTAL"].dropna(), vert=False)
plt.title("Box Plot of Annual Income")
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.boxplot(df["AMT_CREDIT"].dropna(), vert=False)
plt.title("Box Plot of Credit Amount")
plt.show()
```

### Do people with higher income default less?

```python
df.groupby("TARGET")["AMT_INCOME_TOTAL"].describe()
```

```python
plt.figure(figsize=(8,5))

df.boxplot(column="AMT_INCOME_TOTAL", by="TARGET")

plt.title("Income vs Target")
plt.suptitle("")
plt.show()
```

```python
plt.figure(figsize=(8,5))

df.boxplot(column="AMT_INCOME_TOTAL", by="TARGET")

plt.ylim(0, 1000000)   # Zoom to 10 lakh
plt.title("Income vs Target (Zoomed)")
plt.suptitle("")
plt.show()
```
